```python
data = pd.read_csv('/content/train_loan.csv')
```

```python
data.head()
```

```python
# Droping Loan_ID
```

```python
data.drop('Loan_ID',axis = 1)
```

```python
#Removing '+' Sign from Dependents
```

```python
data['Dependents'] = data['Dependents'].str.replace('+','')
```

```python
#Checking for missing values
```

```python
data.isna().sum()
```

```python
#Missing values found at various columns
```

```python
data.dtypes
```

```python
#Missing values handling for categorical data
```

```python
data['Dependents'] = data['Dependents'].fillna(data['Dependents'].mode()[0])
```

```python
#Converting Dependents into Intiger datatype
```

```python
data['Dependents']=data['Dependents'].astype('int')
```

```python
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
```

```python
data['Married'] = data['Married'].fillna(data['Married'].mode()[0])
```

```python
data['Self_Employed'] = data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
```

```python
# Treating Numerical missing values
```

```python
data['LoanAmount'] = data['LoanAmount'].fillna(data['LoanAmount'].median())
```

```python
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].median())
```

```python
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].median())
```

```python
data.isna().sum()
```

```python
#All missing values are treated
```

```python
data.head()
```

```python
#ENCODING
```

```python
x = data.drop('Loan_Amount_Term',axis=1)
y = data['Loan_Amount_Term']
```

```python
x.tail()
```

```python
y.head()
```

```python
#Using One hot encoding
```

```python
x = pd.get_dummies(x,dtype =int)
```

```python
x.head()
```

```python
#Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
y
```

```python
#SCALING
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
```

```python
#Scaling Using Min-max scaler
```

```python
x.loc[:,['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']] = min_max.fit_transform(x[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']])
```

```python
x.head()
```

```python
#TRAINING and TESTING
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

```python
#Importing Regressors RandomForest,XGB,CatBoost
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
```

```python
!pip install catboost
```

```python
from catboost import CatBoostRegressor
```

```python
#Importing Models to variable
```

```python
models = {'Random Forest': RandomForestRegressor(),'XGB Regressor': XGBRegressor(),'Catboost': CatBoostRegressor(verbose=0)}
```

```python
for name,model in models.items():
  model.fit(x_train,y_train)
  print(name+'','trained')
```

```python
#Checking Accuracy of each models
```

```python
from sklearn.metrics import mean_squared_error
```

```python
for name,model in models.items():
  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test,y_pred)
  print(name +' MSE:{:.4f} '.format(mse))
```

```python
#Conclusion : Random forest give maximum accuracy amoung others
```
